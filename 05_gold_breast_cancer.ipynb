# Notebook 05: Camada Gold (Consumo Analítico)

**Objetivo**
- Materializar a camada Gold como **camada de consumo** do pipeline.
- Persistir os dados tratados da Silver sem novas transformações.
- Documentar o **Catálogo de Dados** e a **Linhagem**, uma vez que a Gold
  representa o modelo final para uso analítico.

**Observação**
Neste MVP, a camada Silver já apresenta qualidade e consistência suficientes
para as análises propostas. Dessa forma, a camada Gold atua como uma cópia
dos dados tratados, mantendo aderência ao modelo medalhão e permitindo
futuras evoluções do pipeline.


```python
df_silver = spark.table("silver.breast_cancer")
display(df_silver.limit(10))
```

## Persistência na Camada Gold

Os dados da camada Silver são persistidos na camada Gold **sem alterações**,
caracterizando a Gold como a camada oficial de consumo analítico.

```python
df_silver.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("gold.breast_cancer")

display(spark.table("gold.breast_cancer").limit(10))
```

## Catálogo de Dados — gold.breast_cancer

### Descrição Geral
Conjunto de dados contendo características numéricas extraídas de imagens
de biópsias de câncer de mama, preparado para consumo analítico após etapas
de padronização, validação de qualidade e enriquecimento semântico.

---

### Campos

| Coluna | Tipo | Descrição | Domínio Esperado |
|------|----|---------|----------------|
| mean_radius | double | Média do raio do núcleo celular | > 0 |
| mean_texture | double | Média da textura | > 0 |
| mean_perimeter | double | Média do perímetro | > 0 |
| mean_area | double | Média da área | > 0 |
| mean_smoothness | double | Média da suavidade | > 0 |
| mean_compactness | double | Média da compacidade | > 0 |
| mean_concavity | double | Média da concavidade | ≥ 0 |

## Linhagem dos Dados

- **Fonte Original:**  
  scikit-learn – Breast Cancer Wisconsin (Diagnostic)

- **Coleta:**  
  Carregamento direto via biblioteca Python no Databricks.

- **Staging:**  
  Inspeção inicial dos dados em memória (schema e amostras).

- **Bronze:**  
  Padronização técnica dos nomes das colunas e inclusão de metadados
  de ingestão.

- **Silver:**  
  Validação da qualidade dos dados (completude, unicidade e consistência)
  e enriquecimento semântico da variável alvo.

- **Gold:**  
  Camada de consumo, contendo os dados tratados e prontos para análise.

### Resumo da Camada Gold

- Dados persistidos em formato Delta Lake.
- Nenhuma transformação adicional aplicada.
- Camada destinada exclusivamente ao consumo analítico.
- Catálogo de dados e linhagem documentados neste notebook.
