##❓ Pergunta 1

Qual é a distribuição de casos benignos e malignos no conjunto de dados?

```python
%sql
SELECT
    diagnosis_desc,
    COUNT(*) AS total_registros
FROM gold.breast_cancer
GROUP BY diagnosis_desc
ORDER BY diagnosis_desc;
```

A consulta demonstra a distribuição de registros entre tumores benignos e malignos.
A presença de ambos os grupos viabiliza análises comparativas, validando a
utilidade do conjunto de dados para estudos exploratórios.


##❓ Pergunta 2

Existem diferenças relevantes nas características celulares entre tumores benignos e malignos?


```python
%sql
SELECT
    diagnosis_desc,
    AVG(mean_radius)     AS avg_mean_radius,
    AVG(mean_texture)    AS avg_mean_texture,
    AVG(mean_perimeter)  AS avg_mean_perimeter,
    AVG(mean_area)       AS avg_mean_area,
    AVG(mean_smoothness) AS avg_mean_smoothness
FROM gold.breast_cancer
GROUP BY diagnosis_desc
ORDER BY diagnosis_desc;
```

Os resultados indicam que, em média, tumores malignos apresentam valores mais
elevados em métricas como raio, perímetro e área quando comparados aos tumores
benignos. Essas diferenças sugerem associação entre determinadas características
celulares e o diagnóstico.
