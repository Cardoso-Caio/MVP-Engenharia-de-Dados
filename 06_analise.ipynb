##❓ Pergunta 1

Qual é a distribuição de casos benignos e malignos no conjunto de dados?

```python
%sql
SELECT
  diagnosis_desc AS diagnostico,
  COUNT(*)        AS quantidade,
  ROUND(100.0 * COUNT(*) / SUM(COUNT(*)) OVER (), 2) AS percentual
FROM gold.breast_cancer
GROUP BY diagnosis_desc
ORDER BY quantidade DESC;
```

O conjunto de dados apresenta uma distribuição desbalanceada moderada, com maior quantidade de casos benignos em relação aos malignos.
Esse comportamento é esperado para datasets clínicos reais e não compromete a análise, desde que o desbalanceamento seja conhecido e considerado nas interpretações.

Existem diferenças relevantes nas características celulares entre tumores benignos e malignos?


A análise das médias das **30 variáveis de características celulares**, segregadas por diagnóstico
(benigno e maligno), evidencia que **existem diferenças relevantes em diversas variáveis** do conjunto de dados.

Observa-se que tumores malignos apresentam, em média, valores mais elevados em características
associadas a **tamanho**, **irregularidade** e **complexidade da forma celular**.
Entre as variáveis que se destacam com diferenças mais perceptíveis entre as médias estão:

- `mean_radius`, `mean_perimeter` e `mean_area`, indicando núcleos celulares maiores em tumores malignos;
- `mean_compactness`, `mean_concavity` e `mean_concave_points`, sugerindo maior irregularidade e presença de concavidades;
- variáveis do grupo `worst_*`, como `worst_radius` e `worst_area`, que capturam os cenários mais extremos da morfologia celular.

Por outro lado, algumas variáveis apresentam médias mais próximas entre os dois diagnósticos,
indicando menor poder discriminatório quando analisadas isoladamente.

Essa avaliação global confirma que **nem todas as variáveis contribuem igualmente para a diferenciação**
entre tumores benignos e malignos, justificando a necessidade de uma análise adicional para identificar
quais características apresentam **maior capacidade de diferenciação**, abordada na Pergunta 6.


##❓ Pergunta 6

Existem diferenças relevantes nas características celulares entre tumores benignos e malignos?

```python
%sql
WITH agg AS (
  SELECT
    AVG(CASE WHEN diagnosis = 0 THEN mean_radius END)            AS b_mean_radius,
    AVG(CASE WHEN diagnosis = 1 THEN mean_radius END)            AS m_mean_radius,
    AVG(CASE WHEN diagnosis = 0 THEN mean_texture END)           AS b_mean_texture,
    AVG(CASE WHEN diagnosis = 1 THEN mean_texture END)           AS m_mean_texture,
    AVG(CASE WHEN diagnosis = 0 THEN mean_perimeter END)         AS b_mean_perimeter,
    AVG(CASE WHEN diagnosis = 1 THEN mean_perimeter END)         AS m_mean_perimeter,
    AVG(CASE WHEN diagnosis = 0 THEN mean_area END)              AS b_mean_area,
    AVG(CASE WHEN diagnosis = 1 THEN mean_area END)              AS m_mean_area,
    AVG(CASE WHEN diagnosis = 0 THEN mean_smoothness END)        AS b_mean_smoothness,
    AVG(CASE WHEN diagnosis = 1 THEN mean_smoothness END)        AS m_mean_smoothness,
    AVG(CASE WHEN diagnosis = 0 THEN mean_compactness END)       AS b_mean_compactness,
    AVG(CASE WHEN diagnosis = 1 THEN mean_compactness END)       AS m_mean_compactness,
    AVG(CASE WHEN diagnosis = 0 THEN mean_concavity END)         AS b_mean_concavity,
    AVG(CASE WHEN diagnosis = 1 THEN mean_concavity END)         AS m_mean_concavity,
    AVG(CASE WHEN diagnosis = 0 THEN mean_concave_points END)    AS b_mean_concave_points,
    AVG(CASE WHEN diagnosis = 1 THEN mean_concave_points END)    AS m_mean_concave_points,
    AVG(CASE WHEN diagnosis = 0 THEN mean_symmetry END)          AS b_mean_symmetry,
    AVG(CASE WHEN diagnosis = 1 THEN mean_symmetry END)          AS m_mean_symmetry,
    AVG(CASE WHEN diagnosis = 0 THEN mean_fractal_dimension END) AS b_mean_fractal_dimension,
    AVG(CASE WHEN diagnosis = 1 THEN mean_fractal_dimension END) AS m_mean_fractal_dimension,

    AVG(CASE WHEN diagnosis = 0 THEN radius_error END)           AS b_radius_error,
    AVG(CASE WHEN diagnosis = 1 THEN radius_error END)           AS m_radius_error,
    AVG(CASE WHEN diagnosis = 0 THEN texture_error END)          AS b_texture_error,
    AVG(CASE WHEN diagnosis = 1 THEN texture_error END)          AS m_texture_error,
    AVG(CASE WHEN diagnosis = 0 THEN perimeter_error END)        AS b_perimeter_error,
    AVG(CASE WHEN diagnosis = 1 THEN perimeter_error END)        AS m_perimeter_error,
    AVG(CASE WHEN diagnosis = 0 THEN area_error END)             AS b_area_error,
    AVG(CASE WHEN diagnosis = 1 THEN area_error END)             AS m_area_error,
    AVG(CASE WHEN diagnosis = 0 THEN smoothness_error END)       AS b_smoothness_error,
    AVG(CASE WHEN diagnosis = 1 THEN smoothness_error END)       AS m_smoothness_error,
    AVG(CASE WHEN diagnosis = 0 THEN compactness_error END)      AS b_compactness_error,
    AVG(CASE WHEN diagnosis = 1 THEN compactness_error END)      AS m_compactness_error,
    AVG(CASE WHEN diagnosis = 0 THEN concavity_error END)        AS b_concavity_error,
    AVG(CASE WHEN diagnosis = 1 THEN concavity_error END)        AS m_concavity_error,
    AVG(CASE WHEN diagnosis = 0 THEN concave_points_error END)   AS b_concave_points_error,
    AVG(CASE WHEN diagnosis = 1 THEN concave_points_error END)   AS m_concave_points_error,
    AVG(CASE WHEN diagnosis = 0 THEN symmetry_error END)         AS b_symmetry_error,
    AVG(CASE WHEN diagnosis = 1 THEN symmetry_error END)         AS m_symmetry_error,
    AVG(CASE WHEN diagnosis = 0 THEN fractal_dimension_error END)AS b_fractal_dimension_error,
    AVG(CASE WHEN diagnosis = 1 THEN fractal_dimension_error END)AS m_fractal_dimension_error,

    AVG(CASE WHEN diagnosis = 0 THEN worst_radius END)           AS b_worst_radius,
    AVG(CASE WHEN diagnosis = 1 THEN worst_radius END)           AS m_worst_radius,
    AVG(CASE WHEN diagnosis = 0 THEN worst_texture END)          AS b_worst_texture,
    AVG(CASE WHEN diagnosis = 1 THEN worst_texture END)          AS m_worst_texture,
    AVG(CASE WHEN diagnosis = 0 THEN worst_perimeter END)        AS b_worst_perimeter,
    AVG(CASE WHEN diagnosis = 1 THEN worst_perimeter END)        AS m_worst_perimeter,
    AVG(CASE WHEN diagnosis = 0 THEN worst_area END)             AS b_worst_area,
    AVG(CASE WHEN diagnosis = 1 THEN worst_area END)             AS m_worst_area,
    AVG(CASE WHEN diagnosis = 0 THEN worst_smoothness END)       AS b_worst_smoothness,
    AVG(CASE WHEN diagnosis = 1 THEN worst_smoothness END)       AS m_worst_smoothness,
    AVG(CASE WHEN diagnosis = 0 THEN worst_compactness END)      AS b_worst_compactness,
    AVG(CASE WHEN diagnosis = 1 THEN worst_compactness END)      AS m_worst_compactness,
    AVG(CASE WHEN diagnosis = 0 THEN worst_concavity END)        AS b_worst_concavity,
    AVG(CASE WHEN diagnosis = 1 THEN worst_concavity END)        AS m_worst_concavity,
    AVG(CASE WHEN diagnosis = 0 THEN worst_concave_points END)   AS b_worst_concave_points,
    AVG(CASE WHEN diagnosis = 1 THEN worst_concave_points END)   AS m_worst_concave_points,
    AVG(CASE WHEN diagnosis = 0 THEN worst_symmetry END)         AS b_worst_symmetry,
    AVG(CASE WHEN diagnosis = 1 THEN worst_symmetry END)         AS m_worst_symmetry,
    AVG(CASE WHEN diagnosis = 0 THEN worst_fractal_dimension END)AS b_worst_fractal_dimension,
    AVG(CASE WHEN diagnosis = 1 THEN worst_fractal_dimension END)AS m_worst_fractal_dimension
  FROM gold.breast_cancer
),
diffs AS (
  SELECT
    stack(
      30,

      'mean_radius',               b_mean_radius,               m_mean_radius,
      'mean_texture',              b_mean_texture,              m_mean_texture,
      'mean_perimeter',            b_mean_perimeter,            m_mean_perimeter,
      'mean_area',                 b_mean_area,                 m_mean_area,
      'mean_smoothness',           b_mean_smoothness,           m_mean_smoothness,
      'mean_compactness',          b_mean_compactness,          m_mean_compactness,
      'mean_concavity',            b_mean_concavity,            m_mean_concavity,
      'mean_concave_points',       b_mean_concave_points,       m_mean_concave_points,
      'mean_symmetry',             b_mean_symmetry,             m_mean_symmetry,
      'mean_fractal_dimension',    b_mean_fractal_dimension,    m_mean_fractal_dimension,

      'radius_error',              b_radius_error,              m_radius_error,
      'texture_error',             b_texture_error,             m_texture_error,
      'perimeter_error',           b_perimeter_error,           m_perimeter_error,
      'area_error',                b_area_error,                m_area_error,
      'smoothness_error',          b_smoothness_error,          m_smoothness_error,
      'compactness_error',         b_compactness_error,         m_compactness_error,
      'concavity_error',           b_concavity_error,           m_concavity_error,
      'concave_points_error',      b_concave_points_error,      m_concave_points_error,
      'symmetry_error',            b_symmetry_error,            m_symmetry_error,
      'fractal_dimension_error',   b_fractal_dimension_error,   m_fractal_dimension_error,

      'worst_radius',              b_worst_radius,              m_worst_radius,
      'worst_texture',             b_worst_texture,             m_worst_texture,
      'worst_perimeter',           b_worst_perimeter,           m_worst_perimeter,
      'worst_area',                b_worst_area,                m_worst_area,
      'worst_smoothness',          b_worst_smoothness,          m_worst_smoothness,
      'worst_compactness',         b_worst_compactness,         m_worst_compactness,
      'worst_concavity',           b_worst_concavity,           m_worst_concavity,
      'worst_concave_points',      b_worst_concave_points,      m_worst_concave_points,
      'worst_symmetry',            b_worst_symmetry,            m_worst_symmetry,
      'worst_fractal_dimension',   b_worst_fractal_dimension,   m_worst_fractal_dimension

    ) AS (feature, avg_benign, avg_malign)
  FROM agg
)
SELECT
  feature,
  avg_benign,
  avg_malign,
  ABS(avg_malign - avg_benign) AS abs_diff,
  ROUND(
    100.0 * ABS(avg_malign - avg_benign) / NULLIF(avg_benign, 0),
    2
  ) AS pct_diff_vs_benign
FROM diffs
ORDER BY pct_diff_vs_benign DESC;
```


A partir da ordenação das variáveis pela **diferença percentual entre as médias dos tumores malignos e benignos**, foi possível identificar, de forma objetiva e orientada pelos dados, quais características apresentam maior capacidade de diferenciação entre os diagnósticos.

Os resultados mostram que as **maiores diferenças percentuais** concentram-se em variáveis associadas à
**irregularidade morfológica**, **instabilidade estrutural** e **cenários extremos da forma celular**.
Entre as principais variáveis destacadas pela análise estão:

- **`mean_concavity`**, com diferença percentual superior a **70%**, indicando que tumores malignos apresentam,
  em média, núcleos significativamente mais côncavos do que tumores benignos;
- **`area_error`**, também com diferença superior a **70%**, evidenciando maior variabilidade no tamanho do núcleo
  em tumores malignos;
- **`mean_concave_points`**, com diferença próxima a **70%**, reforçando a presença de bordas mais irregulares
  e complexas nos tumores malignos;
- variáveis do grupo **`worst_*`**, como `worst_concavity`, `worst_area` e `worst_concave_points`,
  que aparecem de forma recorrente entre as maiores diferenças percentuais, refletindo os
  **comportamentos mais extremos** da morfologia celular;
- métricas de erro (`*_error`), como `radius_error` e `perimeter_error`, que indicam maior
  instabilidade e heterogeneidade estrutural nos tumores malignos.

Por outro lado, variáveis relacionadas a textura e suavidade apresentam diferenças percentuais menores,
sugerindo menor poder discriminatório quando analisadas isoladamente.

Essa evidência empírica demonstra que **nem todas as características contribuem igualmente para a diferenciação**
entre tumores benignos e malignos. Em especial, características relacionadas à **concavidade**, **tamanho extremo**
e **variabilidade estrutural** se mostram as mais relevantes neste conjunto de dados.
