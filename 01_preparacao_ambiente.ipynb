## MVP – Engenharia de Dados (Databricks)  
## Notebook 01: Preparação do Ambiente

**Objetivo deste notebook:**  
- Validar o ambiente Databricks (Spark).  
- Criar os schemas (databases) que representam as camadas do pipeline no **modelo medalhão**: `staging`, `bronze`, `silver`, `gold`.

**Observação:**  
A separação por schemas facilita organização, rastreabilidade e avaliação do trabalho.



```python
import pyspark
print("PySpark version:", pyspark.__version__)
```

```python
# Criação das camadas (schemas/databases) do projeto
# staging: área de aterrissagem/preparação (pré-medalhão)
# bronze: dados brutos tabelados (mínimas transformações)
# silver: dados tratados/consistentes
# gold: dados prontos para análise (camada analítica)

spark.sql("CREATE DATABASE IF NOT EXISTS staging")
spark.sql("CREATE DATABASE IF NOT EXISTS bronze")
spark.sql("CREATE DATABASE IF NOT EXISTS silver")
spark.sql("CREATE DATABASE IF NOT EXISTS gold")
```

```python
# Evidência: listar os schemas existentes
spark.sql("SHOW DATABASES").show(truncate=False)
```
